```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("Student_Performance.csv")
df.head()
```

```python
yes=df["Extracurricular Activities"]=="Yes"
no=df["Extracurricular Activities"]=="No"
df.loc[yes,"Extracurricular Activities"]=1
df.loc[no,"Extracurricular Activities"]=0
df
```

```python
column_names=[col for col in df.columns]
```

```python
model=LinearRegression()
model.fit(df[column_names[:-1]],df["Performance Index"])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
test=pd.read_csv("test.csv")
```

```python
yes=test["Extracurricular Activities"]=="Yes"
no=test["Extracurricular Activities"]=="No"
test.loc[yes,"Extracurricular Activities"]=1
test.loc[no,"Extracurricular Activities"]=0
test
```

```python
test["Predicted Performance Index"]=model.predict(test)
test
```
