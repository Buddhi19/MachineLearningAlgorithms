```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
x=np.array([3,5,6,7,8,1,2,3,4,5])
y=np.array([10,23,34,42,70,5,12,10,3,10])
```

```python
plt.scatter(x,y)
```

```python
n=len(y)
learning_rate=0.01
m=0
c=0

for i in range(100):
    y_pred=m*x+c
    cost=(1/n)*sum([value**2 for value in (y-y_pred)])
    plt.scatter(m,cost)

    dm=-(2/n)*sum(x*(y-y_pred))
    dc=-(2/n)*sum(y-y_pred)

    m=m-learning_rate*dm
    c=c-learning_rate*dc

    if(math.isclose(cost,0)):
        break
```

```python
plt.plot(x,m*x+c)
```
